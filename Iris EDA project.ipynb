# Descriptive Statistics

```python
# What is the average, minimum, and maximum values for each of the four numeric 
#features: sepal length, sepal width, petal length, and petal width?
#Sepal_length mean , min, max
sepal_mean=df1["sepal_length"].mean()
sepal_min=df1["sepal_length"].min()
sepal_max=df1["sepal_length"].max()
sepal_mean, sepal_min, sepal_max


```

```python
#sepal width mean, min, max
sepal_width_mean=df1["sepal_width"].mean()
sepal_width_min=df1["sepal_width"].min()
sepal_width_max=df1["sepal_width"].max()
sepal_width_mean, sepal_width_min, sepal_width_max

```

```python
#petal width mean, min, max
petal_width_mean=df1["petal_width"].mean()
petal_width_min=df1["petal_width"].min()
petal_width_max=df1["petal_width"].max()
petal_width_mean, petal_width_min, petal_width_max

```

```python
#petal length mean, min, max
petal_length_mean=df1["petal_length"].mean()
petal_length_min=df1["petal_length"].min()
petal_length_max=df1["petal_length"].max()
petal_length_mean, petal_length_min, petal_length_max
```

```python
#Can you calculate the mean, median, and standard deviation for each feature?
#sepal length mean, median, standard deviation
sepal_L_mean=df1["sepal_length"].mean()
sepal_L_median=df1["sepal_length"].median()
sepal_L_std=df1["sepal_length"].std()
sepal_L_mean,sepal_L_median, sepal_L_std
```

```python
#sepal width mean, median, standard deviation
sepal_W_mean=df1["sepal_width"].mean()
sepal_W_median=df1["sepal_width"].median()
sepal_W_std=df1["sepal_width"].std()
sepal_W_mean,sepal_W_median, sepal_W_std
```

```python
#petal length mean, median, standard deviation
petal_L_mean=df1["petal_length"].mean()
petal_L_median=df1["petal_length"].median()
petal_L_std=df1["petal_length"].std()
petal_L_mean,petal_L_median,petal_L_std
```

```python
#petal width mean, median, standard deviation
petal_W_mean=df1["petal_width"].mean()
petal_W_median=df1["petal_width"].median()
petal_W_std=df1["petal_width"].std()
petal_W_mean,sepal_W_median, sepal_W_std
```

# Data Visualization

```python
#Create a scatter plot to visualize the relationship between sepal length 
# and sepal width. Can you distinguish between the three species based 
#on these two features
sns.scatterplot(x="sepal_length",
               y="sepal_width",
                data=df1,
               hue="species",
               palette="Set1");
plt.title("Distinguish between species", fontsize=23, fontweight="bold");
```

```python
#Plot histograms for petal length and petal width for each species. 
#What insights can you draw from these histograms?
#petal_length
sns.displot(x=df1["petal_length"],
           hue=df1["species"],
            bins= [1,2,3,4,5,6,7],
           kde=True,
           rug=True,
           );
plt.title("Petal_length", fontsize=32,fontweight="bold")
```

```python
#petal_width
sns.displot(x=df1["petal_width"],
           hue=df1["species"],
           kde=True,
           rug=True,
        palette="winter",
            bins=[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5]
           );
plt.title("Petal_width", fontsize=23, fontweight="bold")
```

```python
#Calculate the correlation matrix between the numeric features. Are there strong correlations between any of them?
numeri=df1.select_dtypes(include=["number"])
cor=numeri.corr()
cor
```

```python
#Visualize the correlation matrix as a heatmap for better understanding.
sns.heatmap(cor)
```
